## Oppgave 1a

Det første vi gjør i denne oppgaven er å definere alle konstanter som trengs i programmet.

```python
from pylab import *
q = 3.0
m = 2
dt = 1e-4
t0 = 0
t1 = 1
```

Vi lager så tomme arrays til verdiene vi skal generere, og definerer initialbetingelsene.

```python
t = linspace(t0, t1, (t1 - t0)/dt)

n = len(t)

r = zeros((n,3))
v = zeros((n,3))

r[0] = [0.0, 0.0, 0.0]
v[0] = [0.0, 0.0, 0.0]
E = 5.
```

Så kommer utregninga, vi bruker Euler-Cromers metode:

```python
for i in range(len(t)-1): 
    a = q*E/m
    v[i+1] = v[i] + a*dt
    r[i+1] = r[i] + v[i+1]*dt
```

Så kan vi plotte resulatet:

```python
figure(figsize=(8,6))
plot(t, r[:,0], label="x-direction")
xlabel("Time",fontsize=26)
ylabel("Position",fontsize=26)
legend()
```

## Oppgave 1b

Vi har $F = qE$, vi ønsker å finne et utrykk for posisjonen i x-retning fra $t_{0}=0$ til $t_{1} = 1$.

Vi vet at $F=ma$ fra N2, det gir utrykket:
\begin{equation}
a=\frac{qE}{m}
\end{equation}


Vi kan integrere dette med hensyn på tid:

$v(t)= \int_0^t \frac{qE}{m} = \frac{qE}{m} (t-t_{0})$

Vi vet at $t_{0}=0$ så denne forsvinner. For posisjonen er vi nødt til å integrere igjen med hensyn på tid:

$x(t)= \int_0^t \frac{qE}{m}t = \frac{qE}{2m}t²$


vi kan nå plotte dette inn med det numeriske resulatet:

```python
x_analytical = q*E*t**2/(2*m)
figure(figsize=(8,6))
plot(t, r[:,0], label="Numerical")
plot(t, x_analytical, label="Analytical")
xlabel("Time",fontsize=26)
ylabel("Position",fontsize=26)
legend()
```

Plottet viser bevegelsen til ladningen i E-feltet. Vi ser at de er nesten helt like, det var altså godt nok å bruke Euler-Cromers metode.

```python
from pylab import *
q = 1.0
m = 1.0
dt = 1.0e-4
t0 = 0.0
t1 = 50.0

t = linspace(t0, t1, (t1 - t0)/dt)
n = len(t)

r = zeros((n,3))
v = zeros((n,3))

r[0] = [0.0, 0.0, 0.0]
v[0] = [0.0, 0.0, 0.0]
B = array((0,0,1))

omega = q*norm(B)/m
for i in range(len(t)-1):
    if r[i,0]>-0.1 and r[i,0]<0.1:
        E = cos(omega*t[i])*array([1,0,0]) 
    else:
        E = 0
    a = q*E/m + q/m*cross(v[i],B) 
    v[i+1] = v[i] + a*dt
    r[i+1] = r[i] + v[i+1]*dt
    
    
```

## Oppgave 3b

I denne oppgaven har jeg lagt inn en test for å sjekke om partikkelen har nådd veggen. Om den har gjort det vil B endres til 0.

```python
r[0] = [0.0, 0.0, 0.0]
v[0] = [0.0, 0.0, 0.0]
B = array((0,0,1))

omega = q*norm(B)/m
for i in range(len(t)-1):
    if r[i,0]>-0.1 and r[i,0]<0.1:
        E = cos(omega*t[i])*array([1,0,0]) 
    else:
        E = 0
        
    radius = sqrt(r[i,0]**2+r[i,1]**2)
    if radius>2.6:
        B = array((0,0,0))
        
    a = q*E/m + q/m*cross(v[i],B) 
    v[i+1] = v[i] + a*dt
    r[i+1] = r[i] + v[i+1]*dt
    
```

## Oppgave 3d

For å finne energien til partikkelen i hvert omløp er vi avhengig å se på den kinetiske energien. Vi har at $v=\omega r$ som vi kan sette rett inn i formelen for kinetisk energi: 
\begin{equation}
E_{k}=\frac{1}{2} mv² = \frac{1}{2} m(\omega r)²
\end{equation}
Vi setter videre inn utrykket vårt $\omega$ = qB/m, og forkorter m'ene:
\begin{equation}
\frac{1}{2} \frac{q²B²r²}{m}
\end{equation}
Vi har videre at r = 1m, B = 1T, massen og ladningen til et proton, det gir:
\begin{equation}
\frac{1}{2} \frac{1.602e-19²}{1.67e-27} \approx 48MeV
\end{equation}
Vi regner så ut hvileenergien til elektronet, det gir $E=938$MeV som vi kan sammenlikne med den kinetiske energien $E_{k}/E = 5.11$%


For å kommentere den ikke-relativistiske tilnærmingen gjort i modellen må jeg beregne lorentzfaktoren $\gamma$, vi vet at $E_{k}=48$MeV og kan derfor løse for v i formelen for kinetisk energi. Jeg regnet denne til å bli $v = 9.6\times 10⁷$m/s som kan settes inn i:
\begin{equation}
\gamma = \frac{1}{\sqrt{1-\frac{v²}{c²}}} \approx 1.055
\end{equation}
Det kommer helt an på hvor nøyaktige vi ønsker å være i eksperimentet om dette er en god tilnærming eller ikke.

Vi har den totale tilførte energien lik $E_{k}=48$MeV. Tar vi alt dette i betraktning ser vi at $E_{k}=2\cdot runder \cdot U$

Vi kan løse denne for antall runder og få $ runder = \frac{E_{k}}{2U}$

\begin{equation}
runder = \frac{48}{2\times1\times 1.602e-19} \approx 23982
\end{equation}

Frekvensen til E-feltet blir fra definisjonen $f = 1/T = qB/(2\pi m)$ Hz


Vi setter inn verdiene våre og får:
\begin{equation}
f = \frac{1.602e-19\cdot 1}{2\pi 1.67e-27} \approx 16.2 MHz
\end{equation}
