# <center>Notebook 2: Partikkel i magnetisk felt</center>

I denne oppgåva skal vi rekne ut bana til ein partikkel som beveger seg i eit magnetisk felt. Sjølv om me ikkje har kome til magnetiske felt i pensum heilt enno, så er teorien grei nok: Ein partikkel med ladning $q$ og fart $\vec v$ som er i eit magnetfelt $\vec B$ opplever ei kraft
<br><br>
<center>
$\vec F = q \vec v \times \vec B.$
</center>
<br>
Når me kjenner krafta så finn me akselerasjonen med Newtons andre lov,
<br><br>
<center>
$\vec a = \frac{F}{m} = \frac{q}{m} \vec v \times \vec B.$
</center>



## Steg 1: Last inn pakkar og definér konstantar

Når me løyser fysikk numerisk så får me ikkje til å putte einingane inn i programmet. Det har ikkje noko å seie for python om me arbeider i millimeter eller kilometer, så lenge tala ikkje vert alt for store eller alt for små. I denne oppgåva skal me ikkje bry oss om einingar, men berre la partikkelen ha ladning $q=1$ og masse $m=1$. Viss det hadde vore snakk om eit elektron så ville eininga for ladning dermed ha vore elementærladninga, og eininga for masse ville vore elektronmassen. 

```python
from pylab import *
q = 1.0
m = 1.0
dt = 1e-3
t0 = 0
t1 = 5
```

## Steg 2: Sett opp array og initialbetingingar

Vi sett så opp array som skal lagre tidar og posisjonar for alle tidsstega i simuleringa. I tillegg legg vi inn initialbetingingar.

```python
t = linspace(t0, t1, (t1 - t0)/dt)

n = len(t)
print "Number of elements in t array: ", n

r = zeros((n,3))
v = zeros((n,3))

r[0] = [0.0, 0.0, 0.0]
v[0] = [2.0, 0.0, 0.0]
```

## Steg 3: Velg eit magnetfelt

Me lyt velge oss korleis magnetfeltet skal sjå ut. Lat oss velge eit enkelt felt som er uniformt og peiker opp langs $z$-aksen. (Her kan du leike deg med meir spanande val, om du vil.)

```python
B = array([0.0, 0.0, 5.0])
```

## Steg 4: Rekn ut partikkelbana

Diffferensiallikninga som beskriv partikkelbana er altså
<center>
$\vec a = \frac{F}{m} = \frac{q}{m} \vec v \times \vec B.$
</center>
<br><br>
Ho kan vi løyse numerisk med Euler-Cromers metode, som vi òg brukte i førre notebook:
<br><br>
<center>
$\mathbf{v} (t + \Delta t) = \mathbf v(t) + \mathbf{a}(t) \ \Delta t,$
<br><br>
$\mathbf r(t + \Delta t) = \mathbf r(t) + \mathbf v (t+\Delta t) \ \Delta t.$
</center>
<br><br>
I python-språk blir dette (merk at vi bruker funksjonen `cross()` for å ta kryssproduktet)

```python
for i in range(len(t)-1):
    a   = q/m* cross(v[i],B) 
    v[i+1] = v[i] + a*dt
    r[i+1] = r[i] + v[i+1]*dt
```

## Steg 5: Plott resultat

No er vi klare til å plotte resultata. Først plottar vi $x(t),y(t)$ og $z(t)$ i det same plottet med ulike fargar. $x$-posisjonane er lagra i første kolonne av posisjonsmatrisa `r`, $y$-posisjonane i andre og $z$ i tredje. `legend()`-funksjonen er naudsynt for å liste opp fargane som høyrer til grafane.

```python
figure(figsize=(8,6))
plot(t, r[:,0], label="x-direction")
plot(t, r[:,1], label="y-direction")
plot(t, r[:,2], label="z-direction")
xlabel("Time",fontsize=26)
ylabel("Position",fontsize=26)
legend()
```

Vi kan òg plotte farta på same måte:

```python
figure(figsize=(8,6))
plot(t, v[:,0], label=r"$v_x$")
plot(t, v[:,1], label=r"$v_y$")
plot(t, v[:,2], label=r"$v_z$")
xlabel("Time",fontsize=26)
ylabel("$Positions",fontsize=26)
legend()
```

Lat oss også plotte partikkelbana i 3D! For å gjere det treng me å importere ei ekstra pakke, og dessutan gi den mystiske kommandoen `ax = gca(projection='3d')`, men resten er som tidlegare:

```python
from mpl_toolkits.mplot3d import Axes3D

figure(figsize=(12,10))
ax = gca(projection='3d')
plot(r[:,0], r[:,1], r[:,2])

ax.set_xlabel("$x$",fontsize=26)
ax.set_ylabel("$y$",fontsize=26)
ax.set_zlabel("$z$",fontsize=26)
```
